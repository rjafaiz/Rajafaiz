```python
import pandas as pd
from collections import Counter
import statistics
from scipy.stats import skew, kurtosis

def calculate_mean(data):
    return data.mean()

def calculate_median(data):
    return data.median()

def calculate_mode(data):
    count = Counter(data)
    mode_data = count.most_common(1)
    return mode_data[0][0]

def calculate_std_dev(data):
    return data.std()

def calculate_variance(data):
    return data.var()

def calculate_skewness(data):
    return skew(data)

def calculate_kurtosis(data):
    return kurtosis(data)

df_data = pd.read_csv(r'C:\Users\hp\Downloads\heart_disease_uci.csv')

for column in df_data.select_dtypes(include='number').columns:
    data_column = df_data[column].dropna() 
    
    mean = calculate_mean(data_column)
    median = calculate_median(data_column)
    mode = calculate_mode(data_column)
    std_dev = calculate_std_dev(data_column)
    variance = calculate_variance(data_column)
    skewness = calculate_skewness(data_column)
    kurt = calculate_kurtosis(data_column)
    
    print(f"Column: {column}")
    print(f"  Mean: {mean}")
    print(f"  Median: {median}")
    print(f"  Mode: {mode}")
    print(f"  Standard Deviation: {std_dev}")
    print(f"  Variance: {variance}")
    print(f"  Skewness: {skewness}")
    print(f"  Kurtosis: {kurt}")
    print()
```

Pada nilai korelasi pada setiap variable data numerik dan menemukan data yang memiliki korelasi tertinggi Nilai 1 menunjukkan adanya korelasi sempurna positif, yang berarti bahwa ketika satu variabel meningkat, yang lain juga meningkat secara proporsional; nilai -1 menunjukkan adanya korelasi sempurna negatif, yang berarti bahwa ketika satu variabel meningkat, yang lain turun secara proporsional; dan nilai 0 menunjukkan bahwa tidak ada korelasi linear antara dua variabel.
Pasangan variabel dengan korelasi tertinggi sering ditemukan dalam dataset yang berkaitan dengan penyakit jantung yang Anda sebutkan. Pasangan variabel ini biasanya terdiri dari num (penanda keparahan penyakit jantung) dan variabel klinis lainnya yang relevan, seperti ca (jumlah pembuluh darah utama berwarna dengan fluoroskopi) atau oldpeak (depresi ST yang disebabkan oleh latihan dibandingkan dengan tidur). Misalnya, mungkin ada korelasi tertinggi antara num dan ca. Ini karena ca adalah ukuran langsung tingkat keparahan kondisi kardiovaskular, dan nilai num mungkin sangat berkorelasi dengan nilai ca.

Nilai num dan ca terjadi pada berbagai tingkat, tetapi plot ini tidak menunjukkan hubungan linear yang jelas. Dengan kata lain, penurunan nilai num.

Hampir setiap tingkat num menunjukkan beberapa nilai ca, menunjukkan bahwajumlah pembuluh darah (ca) dapat sangat berbeda dari tingkat keparahan penyakit jantung (num).
